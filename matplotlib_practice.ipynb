```python
# Import dependencies.
import matplotlib.pyplot as plt
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# Create the plot
plt.plot(x_axis, y_axis)
```

```python
# using object-oriented interface
# Create the plot with ax.plt()

fig, ax = plt.subplots()
ax.plot(x_axis, y_axis)
```

```python
# same code could also be written as 
# Create the plot with ax.plt()
fig = plt.figure()
ax = fig.add_subplot()
ax.plot(x_axis, y_axis)
```

```python
# to run a quick graph, we can also write it as 
# Create the plot with ax.plt()
ax = plt.axes()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot.
plt.plot(x_axis, y_axis)
plt.show()
```

```python
# Create the plot and add a label for the legend.
plt.plot(x_axis, y_axis,linewidth= '2', label='Boston', marker= 'D',color = 'green')
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Set the y limit between 0 and 45.
plt.ylim(0, 45)
# Create a title.
plt.title("PyBer Fare by Month")
# Add the legend.
plt.legend()
# add gridlines to the graph
plt.grid()
```

```python
# Scatter plot - 5.1.7
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
plt.plot(x_axis, y_axis, 'o')
```

```python
plt.scatter(y_axis, x_axis, color ='r')
plt.title("PyBer Fare by Month")
plt.legend('Chicago')
plt.xlabel("Fare($)")
plt.ylabel("Date")
# invert y axis
plt.gca().invert_yaxis()
# set limit to x axis
plt.xlim(0,50)
```

```python
plt.scatter(x_axis, y_axis, s=y_axis)
```

```python
y_axis_larger = []
for data in y_axis:
  y_axis_larger.append(data*3)
plt.scatter(x_axis, y_axis, s=y_axis_larger)
```

```python
# create a bubble chart using MATLAB method
plt.scatter(x_axis, y_axis)
```

```python
# refactored from above
plt.scatter(x_axis, y_axis, s = [i * 3 for i in y_axis])
```

```python
# Create a Scatter Plot Using the Object-Oriented Interface
fig, ax = plt.subplots()
ax.scatter(y_axis, x_axis, color='xkcd:sky blue', edgecolor = 'BLACK',alpha = .80 ,s=[i * 5 for i in y_axis])
ax.invert_yaxis()
plt.title('PyBer Fare by Month')
plt.legend('Boston')
plt.xlabel("Fare($)")
plt.ylabel("Date")

# set limit to x axis
plt.xlim(0,50)
```

```python
plt.pie(y_axis, labels=x_axis)
plt.show()
```

```python
plt.subplots(figsize=(8, 8))
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
plt.pie(y_axis, explode=explode_values, labels=x_axis, autopct='%.1f%%')
```
